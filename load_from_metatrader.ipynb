```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import MetaTrader5 as mt
from datetime import datetime, timedelta
import pytz
```

```python
if not mt.initialize():
    print("initialize() failed")
    mt.shutdown()

print(mt.terminal_info())
print(mt.version())
```

```python
timezone = pytz.timezone("Etc/UTC")
date_from = datetime(2011, 1, 1, tzinfo=timezone)
date_to = datetime(2020, 1, 1, tzinfo=timezone)
candles = mt.copy_rates_range('EURUSD', mt.TIMEFRAME_M5, date_from, date_to)
print(mt.last_error())
```

```python
def to_dataframe(candles, to_date=False):
    df = pd.DataFrame(data=candles, columns=['time', 'open', 'high', 'low', 'close', 'tick_volume', 'spread', 'real_volume'])
    if to_date:
        df['time'] = df['time'].apply(datetime.utcfromtimestamp)
    return df

data = to_dataframe(candles)
data.head()
```

```python
data.to_parquet('resources/EURUSD_2011-2020_M5.parquet')
```
